```python
import pandas as pd
import numpy as np
```

```python
main_df = pd.read_csv("https://raw.githubusercontent.com/4GeeksAcademy/naive-bayes-project-tutorial/main/playstore_reviews.csv")
```

```python
main_df.head()
```

```python
main_df["review"] = main_df["review"].str.strip().str.lower()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

from sklearn.model_selection import train_test_split
```

```python
X = main_df[['review']]
y = main_df[['polarity']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
vec_model = CountVectorizer(stop_words = "english")
X_train_vec = vec_model.fit_transform(X_train['review']).toarray()
X_test_vec = vec_model.transform(X_test['review']).toarray()
```

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
```

```python
gnb_model = GaussianNB()

gnb_model.fit(X_train_vec, y_train)
```

```python
gnb_train_preds = gnb_model.predict(X_train_vec)
gnb_test_preds = gnb_model.predict(X_test_vec)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
print(f"Confusion matrix: {confusion_matrix(y_train, gnb_train_preds)}")
print(f"Accuracy score: {accuracy_score(y_train, gnb_train_preds)}")
print(f"Classification report: {classification_report(y_train, gnb_train_preds)}")
```

```python
b_model = BernoulliNB()

b_model.fit(X_train_vec, y_train)
```

```python
b_train_preds = gnb_model.predict(X_train_vec)
b_test_preds = gnb_model.predict(X_test_vec)
```

```python
print(f"Confusion matrix: {confusion_matrix(y_train, b_train_preds)}")
print(f"Accuracy score: {accuracy_score(y_train, b_train_preds)}")
print(f"Classification report: {classification_report(y_train, b_train_preds)}")
```
